```python
#loading document and performing ocr using unstructured
from langchain.document_loaders import UnstructuredPDFLoader

filename='/content/CompulsoryGuidelinesForSafeTravel.pdf'
loader=UnstructuredPDFLoader(filename)
docs=loader.load_and_split()

for doc in docs:
  print(doc.page_content)
```

```python
docs
```

```python
#perform chunking on the extracted text
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=100)
split_docs=text_splitter.split_documents(docs)
```

```python
#removing \n characters from text
from langchain.schema import Document

final_docs1=[]
for doc in split_docs:
  final_docs1.append(Document(page_content=doc.page_content.replace("\n",""),
                          metadata={'source':filename}
                          ))
```

```python
#replacing the '|' wrongly parsed as I
final_docs=[]
for doc in final_docs1:
  final_docs.append(Document(page_content=doc.page_content.replace("|","I"),
                          metadata={'source':filename}
                          ))
```

```python
#final document chunks
final_docs
```

```python
#initiating embeddings model and llm from openai
from langchain_openai import ChatOpenAI
from langchain_openai import OpenAIEmbeddings
from google.colab import userdata
llm=ChatOpenAI(model='gpt-3.5-turbo',api_key=userdata.get('openai-key'),temperature=0)
embeddings=OpenAIEmbeddings(api_key=userdata.get('openai-key'))
```

```python
#initializing faiss index and vectorstore
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_community.vectorstores import FAISS

index = faiss.IndexFlatL2(len(embeddings.embed_query("hello world")))

vector_store = FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
#assigning unique ids to each document chunk and inserting them into the vectorstore
from uuid import uuid4
uuids = [str(uuid4()) for _ in range(len(final_docs))]
vector_store.add_documents(documents=final_docs,ids=uuids)
```

```python
#creating a retrieval chain using langchain prompts, retriever and chains to pass a query and generate a response from the llm
from langchain_community.chat_models import ChatOpenAI
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain.chains import create_retrieval_chain
from langchain import hub

retrieval_qa_chat_prompt = hub.pull("langchain-ai/retrieval-qa-chat")
# llm = ChatOpenAI()
retriever = vector_store.as_retriever(search_type="mmr", search_kwargs={"k": 1})
combine_docs_chain = create_stuff_documents_chain(
    llm, retrieval_qa_chat_prompt
)
retrieval_chain = create_retrieval_chain(retriever, combine_docs_chain)

retrieval_chain.invoke({"input": "what do i need to declare before travel"})
```

```python
#inplementing simple ui using gradio to make chatbot user friendly.
import gradio as gr
inputs=gr.Textbox(label="enter your query")
outputs=gr.Textbox(label="response")
def generate_response(query: str):
  return retrieval_chain.invoke({'input':query})['answer']

```

```python
chatbot=gr.Interface(fn=generate_response, inputs=inputs, outputs=outputs)
chatbot.launch()
```
